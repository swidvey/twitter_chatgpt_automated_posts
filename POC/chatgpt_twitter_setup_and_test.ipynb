# Putting It All Together

Now, the moment you've all been waiting for...we're finally bringing our friendly ChatGPT bot to Twitter! It's like introducing your two best friends to each other and watching them hit it off immediately. Except in this case, it's all virtual, and the conversation is a little one-sided. But hey, it's still pretty cool! 
<br>
<br>
The "text-davinci-003" engine is one of the models available from OpenAI. It's capable of generating highly creative and coherent text across a wide range of topics, and has been trained on a massive amount of data to enable it to understand language in a highly nuanced way. I found the output better than some of the other engine choices.

```python
# Already imported above
# import openai
# import tweepy

# Set your API key
openai.api_key      = "-- your key -- "
consumer_key        = "-- your key -- "
consumer_secret     = "-- your key -- "
access_token        = "-- your key -- "
access_token_secret = "-- your key -- "


# Send a prompt to the model and receive a response
prompt = "Give me a tweet about chatgpt and how I developed Twitter API link. It should say that this post was created by chatgpt"
response = openai.Completion.create(
    engine="text-davinci-003",
    prompt=prompt,
    temperature=0.7,
    max_tokens=256,
    n=1,
    stop=None,
    frequency_penalty=0,
    presence_penalty=0
)


print(response.choices[0].text)
```

You could rerun this changing the prompts until you find a tweet you like. 
Then we send the responce to twitter! 

```python
# Authenticate with Tweepy
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)


tweet = response.choices[0].text
api.update_status(tweet)
```
